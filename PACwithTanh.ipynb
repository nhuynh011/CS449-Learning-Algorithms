Design a Boolean classifier (outputs +1/-1) which is (approximately) accurate on the training set. This means that the total number of misclassifications that the classifier makes on the training set should be (near) zero. Use a classifier that is a neural network with one output and one layer of hidden units.

```python
from os import error
import numpy as np
from sklearn.preprocessing import normalize
# Load data
x = np.genfromtxt('/content/a2-train-data.txt', delimiter=' ')
y = np.genfromtxt('/content/a2-train-label.txt', delimiter=' ')

rows, cols = x.shape

# example of the xavier weight initialization
from math import sqrt
# number of nodes in the previous layer
n = 10
# calculate the range for the weights
lower, upper = -(1.0 / sqrt(n)), (1.0 / sqrt(n))
# generate random numbers
numbers = rand(1000)
# scale to the desired range
scaled = lower + numbers * (upper - lower)
# summarize
print(lower, upper)
print(scaled.min(), scaled.max())
print(scaled.mean(), scaled.std())

# Training response vector:
yhat = [0]*rows

# Error counter
err = 0
```

Forward function (finish first before running back)

```python
# Calculate yhat, forward pass
def forwpass(error, w):
  for i in range(rows):
      tempw = np.dot(w, x[i]) # -> reduces layer to 1 input

      # Activate tanh()
      yhat[i] = np.tanh(tempw)

      # Calculate total error
      error += (yhat[i] - y[i])**2
      return error
```

Backwards function for gradient
update w like so <- w- learnRate * slope of error (dE(w)/dw)

sig(j-1) = sig'(z(j-1)) * ∑(sig(z(j-1)) - y)
         = yhat[j-1] * ∑(yhat[j-1] - y[j-1])

```python
# Back propagation:
def backprop(c, w):
  for i in range(rows):
      tempe = 1 - (np.tanh(yhat[i]))**2 # Compute slope of error for all i
      w = w - c*(1-(np.tanh(w)**2)) # Update weights for each trial of x
  return w
```

Layer with n nodes, decide later.

Loop:

```python
acc = 0.95
c = 0.02 # learning rate

while True:
  new_err = forwpass(err, w)
  err = new_err
  print(backprop(c, w))
  if err<((1-acc)*cols):
    break
```
