Load x, initialize weights, error, label vector

```python
import numpy as np
from sklearn.preprocessing import normalize
# Load data
x = np.genfromtxt('/content/train-a1-449.txt', dtype = str, delimiter=' ')

rows, cols = x.shape

# Label vector:
y = [-1] * rows
for i in range(rows):
    if x[i, 1024] == 'Y':
        y[i] = 1

# Remove y from x
x = np.delete(x, -1, axis=1)

# Normalize x
x = normalize(x, norm = "l2")
rows, cols = x.shape

# Weight vector
w = [0]*1024

# Predictor vector
yhat = np.zeros(rows)

# Min for margin of error
m = 1000000

# Full w
wfull = ''
```

Activation

```python
def sign(x):
  return 1 if x >= 0 else -1
```

Min function for margin

```python
def min(x, y):
  return x if x <= y else y
```

Apply

```python
from re import M
while True:
  err = 0
  for i in range(rows):
    yhat[i] = sign(np.dot(w, x[i]))
    if sign(np.dot(w, x[i])) * y[i] <=0 : # = 1 if both agree, = 0 or -1 if don't agree -> mistake
      tempw = w
      w = w + y[i] * x[i]

      # Error counter
      err += 1

  print(err)
  if err == 0:
    break

for i in range(rows):
  temp = np.absolute(np.dot(tempw, x[i])/(np.sqrt(np.dot(x[i], x[i]))))
  temp2 = np.absolute(np.dot(w, x[i])/(np.sqrt(np.dot(x[i], x[i]))))
  current = min(temp, temp2)
  m = min(current, m)

print("w vector")
for i in range(len(w)):
  wfull += str(w[i]) + ", "

print(wfull)
print("\nMargin", m)
```
